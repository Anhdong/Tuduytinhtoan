```python
#W5A3
from math import sqrt
n = int(input())

def is_prime(n):
    ds = [1]*int(sqrt(n)+1)
    for i in range(2,int(sqrt(n))+1):
        print(i)
        if ds[i] == 1:
            if n%i==0:
                return False
            temp = i
            while temp <= int(sqrt(n)):
                ds[temp] = 0
                temp += i     
    return True

print(is_prime(n))

```

```python
#W5A4
from math import sqrt
n = int(input())
def is_perfect_number(n):
    total = 0
    for i in range(1,int(sqrt(n)+1)):
        if n % i == 0:
            total += i
    return True if total == n else False

print(is_perfect_number(n))
```

```python
#W5A5
int_list = list(map(int,input().split()))
k = int(input())

def find_pos(list,k):
    index = 0
    for i in list:
        index += 1
        if i == k:
            return index
    return -1
print (find_pos(int_list,k))
```

```python
#W5A6
def factorial(n):
    product = 1
    for i in range(2,n+1):
        product *= i
    return product

print(factorial(6))
```
