```python
ls1 = list(map(int,input().split()))
ls2 = list(map(int,input().split()))
len1, len2= len(ls1),len(ls2)

c1,c2 = 0,0
merged_ls = []

while c1!=len1 and c2!=len2:
    if ls1[c1] <= ls2[c2]:
        merged_ls.append(ls1[c1])
        c1 +=1
    else:
        merged_ls.append(ls2[c2])
        c2 +=1 
while c1 != len1:
    merged_ls.append(ls1[c1])
    c1 +=1
while c2 != len2:
    merged_ls.append(ls2[c2])
    c2 +=1

print(merged_ls)
```
