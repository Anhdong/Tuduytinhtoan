```python
#W4A8
n = int(input("Nhap so nguyen duong: "))
steps = 0

def is_palindrome(n):
    half_n = len(str(n))//2
    if len(str(n)) % 2 == 0:
        return str(n)[:half_n] == str(n)[:half_n-1:-1]  
    else:
        return str(n)[:half_n] == str(n)[:half_n:-1]

while not is_palindrome(n):
    n += int(str(n)[::-1])
    steps += 1

print(steps)
print(n) 
```

```python
#W4A9
n = int(input("Nhap so nguyen duong: "))
number_list = []

def has_duplicates(x):
    x_list = list(str(x))
    return len(x_list) != len(set(x_list))

#danh sach so chinh phuong
for i in range(1,n+1):
    if i**0.5%1==0 and not has_duplicates(i):
        number_list.append(i)

if number_list == []:
    print("No numbers")
else:
    for i in number_list:
        print(i,end=" ")
```

```python
#W4A10
'''Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài
dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x
L). Nếu có nhiều số cùng độ dài, chọn số nhỏ nhất.'''

n = int(input("Nhap so nguyen n: "))


#check even or odd
def is_even(x): return x % 2 == 0

#tra ve do dai day callatz cua z
def collatz_length(x):
    length = 1
    while x != 1:
        length += 1
        if is_even(x):
            x /= 2
        else:
            x = x*3 +1
    return length

#goi s la so can tim va s_L la do dai cua day
s = 0
s_L = 0

for i in range(1,n+1):
    if collatz_length(i) > s_L:
        s_L = collatz_length(i)
        s = i

print(s , s_L)


```
