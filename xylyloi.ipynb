```python
def sum_power(n:int):
    total = 0
    for i in str(n):
        total += pow(int(i),2)
    return total

try:
    n = int(input())
except ValueError:
    print("n khong la so")
else:
    if n == 1: print("YES")
    else:
        my_list=[]
        my_list.append(n)
        while True:
            if sum_power(my_list[-1]) ==1: 
                print("YES")
                break
            elif sum_power(my_list[-1]) in my_list:
                print("NO")
                break
            my_list.append(sum_power(my_list[-1]))

```
