```python
#W6A9

#delete unecessary syntax
def del_space(word):
#Find spaces and commas
    syntax_list = []
    for i in range(0,len(word)):
        if word[i] == " " or word[i] == "'":
            syntax_list.append(i)
#Remove spaces and commas
    for i in syntax_list[::-1]:
        word = word[:i] + "" + word[i+1:]

    return word


#receive dictionary inputs
def dict_input():
    clean_list = del_space(input()[1:-1]).split(',')
    my_dict = {}

    for i in clean_list:
        key, value = i.split(":")

        if key not in my_dict:
            my_dict[key] = int(value)
        else:
            my_dict[key] += int(value)
    
    return my_dict

#receive 2 dictionaries
a = dict_input()
b = dict_input()

#adding
for key, value in a.items():
    if key not in b:
        b[key] = value
    else:
        b[key] += value

#sort dict
new_dict = {}
for i in sorted(b.keys()):
    new_dict[i] = b[i] 
    
print(new_dict)



```

```python
#W6A12

num_list = list(map(int,input("Nhap day so").split()))
large_num = (0,0) #(number,amount)
flags = [0] * len(num_list)

for i in range(len(num_list)):
    counter = 0
    if flags[i] == 0:
        flags[i] = 1
        counter += 1
        for j in range(i+1,len(num_list)):
            if num_list[j] == num_list[i] and flags[j] == 0:
                flags[j] = 1
                counter += 1
    if counter >= large_num[1]:
        large_num = (num_list[i],counter)
    if counter == large_num[1] and num_list[i] < large_num[0]:
        large_num = (num_list[i],counter)

print(large_num[0])
```

```python
#W6A15
input_list = list(input().split())
k = int(input())
my_dict = {}

def is_num(string):
    for i in range(0,9):
        if str(i) in string:
            return True
    return False

for i in range(len(input_list)):
    if not is_num(input_list[i]):
        if int(input_list[i+1]) > k:
            my_dict[input_list[i]] = input_list[i+1]


print(my_dict)
```

```python
#W6A17
n = int(input("size: "))
matrix = []

for i in range(n):
    row = []
    for j in range(n):
        row.append(input(f"element in ({i+1},{j+1}): "))
    matrix.append(row)

main_diagonal = []
second_diagonal = []

for i in range(n):
    main_diagonal.append(matrix[i][i])
    second_diagonal.append(matrix[i][n-1-i])

print(main_diagonal)
print(second_diagonal)
```

```python
#W6A21
n = list(input().split())
my_dict = {}
flags = [0]*len(n)

for i in range(len(n)):
    if flags[i] == 0:
        flags[i] = 1
        temp_list = []

        temp_list.append(i)

        for j in range(i,len(n)):
            if flags[j] == 0 and n[j] == n[i]:
                flags[j] = 1
                temp_list.append(j)

        #add position to dict
        my_dict[n[i]] = tuple(temp_list)

print(my_dict)
```
