```python
#W8A1
r, h = map(int,input().split())


class Cylinder:
    pi = 3.14
    def __init__(self,radius,height):
        self.radius = radius
        self.height = height

    def total_area(self):
        return self.radius*self.height*self.pi*2
    
    def volume(self):
        return self.height*self.pi*self.radius**2

cyn1= Cylinder(r,h)

print(cyn1.total_area())
```

```python
#W8A2
d,m,y = map(int,input().split("/"))

class Date:
    def __init__(self,day,month,y):
        self.day = d
        self.month = m
        self.year = y

    def is_leap(self):
        if not self.year % 400:
            return True
        if self.year % 100:
            if not self.year % 4:
                return True
        return False

    #create a month_dict with month as key and number of days as val
    list_30 = ['4','6','9','11']
    list_31 =['1','3','5','7','8','10','12'] 

    def month_dict(self):
        return {**{key:30 for key in self.list_30},**{key:31 for key in self.list_31},'2': 29 if self.is_leap() else 28}



    def is_valid(self):
        if self.year < 0: return "INVALID"
        if self.day > self.month_dict()[str(self.month)]:
            return "INVALID"
        return "VALID"
        
    def next_date(self):
        if self.is_valid() == "INVALID": return "Date is invalid!!!"
        if self.day == self.month_dict()[str(self.month)]:
            if self.month == 12:
                self.month, self.day = 1,1
                self.year += 1

            else:
                self.day = 1
                self.month += 1
        else:
            self.day += 1
    
        return f"{self.day}/{self.month}/{self.year}"
        
date1 = Date(d,m,y)
print(date1.is_leap())
print(date1.is_valid())
print(date1.next_date())


        



    


```
