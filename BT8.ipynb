```python
#W8A1
r, h = map(int,input().split())


class Cylinder:
    pi = 3.14
    def __init__(self,radius,height):
        self.radius = radius
        self.height = height

    def total_area(self):
        return self.radius*self.height*self.pi*2
    
    def volume(self):
        return self.height*self.pi*self.radius**2

cyn1= Cylinder(r,h)

print(cyn1.total_area())
```
