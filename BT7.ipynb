```python
#W7A1
my_array = list(map(int,input().split()))
x = int(input())

def binary_search(arr,target): 
    
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = int((left + right) /2)
        if arr[mid] == target:
            return mid
        elif arr[mid] >= target:
            right = mid -1
        elif arr[mid] <= target:
            left = mid + 1
    
    return -1

print(binary_search(my_array,x))
```

```python
#W7A3
my_array = list(map(int,input().split()))

def bubble_sort(arr):
    for i in range(0,len(arr)):
        swapped = False
        for j in range(0,len(arr)-i-1):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1] = arr[j+1],arr[j]
                swapped = True
        if swapped == False:
            break
    return arr

def selection_sort(arr):
    for i in range(len(arr)):
        min_index = i
        for j in range(i+1,len(arr)):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[min_index],arr[i] = arr[i],arr[min_index]
    return arr

print(selection_sort(my_array))
```

```python
#W7A8
def get_interval():
    n = list(map(int,input().split()))
    nested_array = []
    for i in range(len(n)):
        if i%2 == 0:
            nested_array.append([n[i]])
        else:
            nested_array[(i-1)//2].append(n[i])
    return nested_array


my_intervals = get_interval()
my_querries = list(map(int,input().split()))

print(my_intervals)
print(my_querries)



for i in range(len(my_querries)):
    has_len = False
    best_length =0
    for j in range(len(my_intervals)):
        if my_querries[i] >= my_intervals[j][0] and my_querries[i] <= my_intervals[j][1] and best_length > my_intervals[j][1]-my_intervals[j][0]+1 and has_len == True:
            best_length = my_intervals[j][1]-my_intervals[j][0]+1
            
        elif has_len == False and my_querries[i] >= my_intervals[j][0] and my_querries[i] <= my_intervals[j][1]:
            has_len = True
            best_length = my_intervals[j][1]-my_intervals[j][0]+1
    if has_len == False: best_length = -1
    my_querries[i] = best_length

print(my_querries)
```

```python
#W7A10
my_arr = list(map(int,input("Nhap day so").split()))
m = int(input("Nhap so nguyen m"))

sum_prefix = []

sum_prefix.append(0)

ans = 0

for i in range(0,len(my_arr)):
    sum_prefix.append(sum_prefix[i]+my_arr[i])

for i in range(len(sum_prefix)):
    for j in range(i,len(sum_prefix)):
        temp = (sum_prefix[j]-sum_prefix[i])%m
        if temp > ans: ans = temp

print(ans)
```
